# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type.

total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each city type

total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 

average_fare_per_ride = total_fares / total_rides
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 

average_fare_per_driver = total_fares / total_drivers
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 

pyber_summary_df = pd.DataFrame({
    "Total Rides": total_rides,
    "Total Drivers": total_drivers,
    "Total Fares": total_fares,
    "Average Fare per Ride": average_fare_per_ride,
    "Average Fare per Driver": average_fare_per_driver
})

pyber_summary_df.head()
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.

pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,.0f}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,.0f}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df.head()
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() on the pyber_data_df to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

pyber_by_date_df = pd.DataFrame(pyber_data_df.groupby(["date","type"]).sum()["fare"])

#urban_fare = pyber_data_df[(pyber_data_df["type"] == "Urban")].groupby(["date"]).sum()["fare"]
#suburban_fare = pyber_data_df[(pyber_data_df["type"] == "Suburban")].groupby(["date"]).sum()["fare"]
#rural_fare = pyber_data_df[(pyber_data_df["type"] == "Rural")].groupby(["date"]).sum()["fare"]

# Combine each Series for sum of fares for each date into single DataFrame.
#pyber_by_date_df = pd.DataFrame({"Rural":rural_fare,
#                                 "Suburban":suburban_fare,
#                                 "Urban":urban_fare})
pyber_by_date_df.head(20)
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

pyber_by_date_df = pyber_by_date_df.reset_index()
pyber_by_date_df.head()
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

pyber_by_date_df = pyber_by_date_df.pivot(index="date", columns="type", values="fare")
pyber_by_date_df.head(20)
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2018-01-01':'2018-04-29'.

pyber_by_date_pivot_df = pd.DataFrame(pyber_by_date_df.loc['2019-01-01':'2019-04-30'])
pyber_by_date_pivot_df.tail()
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

pyber_by_date_pivot_df.index = pd.to_datetime(pyber_by_date_pivot_df.index)
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
pyber_by_date_pivot_df.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_resample_df = pd.DataFrame()

pyber_resample_df["Rural"] = pyber_by_date_pivot_df.Rural.resample('W').sum()
pyber_resample_df["Suburban"] = pyber_by_date_pivot_df.Suburban.resample('W').sum()
pyber_resample_df["Urban"] = pyber_by_date_pivot_df.Urban.resample('W').sum()
pyber_resample_df = pd.DataFrame(pyber_resample_df.loc['2019-01-01':'2019-04-30'])

pyber_resample_df.tail(20)
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Plot resampled pyber data by city type using df.plot() function
ax = pyber_resample_df.plot(figsize=(30,10))

ax.set_title("Total Fare by City Type", fontsize = "30")
ax.set_xlabel("", fontsize = "20")
ax.set_ylabel("Fare ($USD)", fontsize = "20")
ax.grid(True)

ax.legend(["Rural","Suburban", "Urban", ], title="type", loc="center", title_fontsize="20", fontsize="20", mode="Expanded")

plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
